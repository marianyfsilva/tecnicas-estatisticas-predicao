```r
# Importação da biblioteca padrão ggplot2
library(ggplot2) 
```

```r
# Leitura dos dados
csv_content <- read.csv("DADOS-CRIMES-UCI.csv")

print(paste("O número de colunas é: ", ncol(csv_content))) # deve ser 7
print(paste("O número de linhas é: ", nrow(csv_content)))  # deve ser 2128

print(paste("O número de colunas é: ", ncol(csv_content)))
print(paste("O número de linhas é: ", nrow(csv_content)))

data = data.frame(
    # Variável resposta
    CRIMES = log(csv_content["CRIMES"]),
    # Variaveis explicativas
    POPULACAO = log(csv_content["POPULACAO"]),
    RENDA = log(csv_content["RENDA"]),
    SEM_BASICO = csv_content["SEM_BASICO"],
    SEM_MEDIO = csv_content["SEM_MEDIO"],
    COM_SUPERIOR = csv_content["COM_SUPERIOR"],
    DESEMPREGADOS = csv_content["DESEMPREGADOS"]
)

# imprimir início do data frame
head(data)
```

```r
print("Diagrama de dispersão")
print("[Populacao] X [Total de crimes violentos e não violentos]")

g1 <- ggplot(data=data, aes(x=POPULACAO, y=CRIMES))+
  geom_point() +
  labs(x = 'Populacao', y = 'Total de crimes violentos e não violentos') +
  theme_minimal()
g1 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Indice de Renda per capta] X [Total de crimes violentos e não violentos]")

g2 <- ggplot(data=data, aes(x=RENDA, y=CRIMES))+
  geom_point() +
  labs(x = 'Indice de Renda per capta', y = 'Total de crimes violentos e não violentos') +
  theme_minimal()

g2 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Pessoas com nível básico incompleto] X [Total de crimes violentos e não violentoss]")

g3 <- ggplot(data=data, aes(x=SEM_BASICO, y=CRIMES))+
  geom_point() +
  labs(x = '% Pessoas com nível básico incompleto', y = 'Total de crimes violentos e não violentoss') +
  theme_minimal()
g3 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Pessoas com nível médio incompleto] X [Total de crimes violentos e não violentos]")

g4 <- ggplot(data=data, aes(x=SEM_MEDIO, y=CRIMES))+
  geom_point() +
  labs(x = '% Pessoas com nível médio incompleto', y = 'Total de crimes violentos e não violentos') +
  theme_minimal()

g4 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Pessoas com nível superior completo] X [Total de crimes violentos e não violentos]")

g5 <- ggplot(data=data, aes(x=COM_SUPERIOR, y=CRIMES))+
  geom_point() +
  labs(x = '% Pessoas com nível superior completo', y = 'Total de crimes violentos e não violentos') +
  theme_minimal()

g5 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Desempregados] X [% Roubos / 100 mil habitantes]")

g6 <- ggplot(data=data, aes(x=DESEMPREGADOS, y=CRIMES))+
  geom_point() +
  labs(x = '% Desempregados', y = 'Total de crimes violentos e não violentos') +
  theme_minimal()

g6 + geom_smooth(method = 'lm', se = F)
```

```r
# Correlação, Variância e Covariância

print("POPULACAO")
cor(data$POPULACAO, data$CRIMES)
var(data$POPULACAO, data$CRIMES, na.rm = FALSE, use)
cov(data$POPULACAO, data$CRIMES, use = "everything",
    method = c("pearson", "kendall", "spearman"))

print("RENDA")
cor(data$RENDA, data$CRIMES)
var(data$RENDA, data$CRIMES, na.rm = FALSE, use)
cov(data$RENDA, data$CRIMES, use = "everything",
    method = c("pearson", "kendall", "spearman"))

print("SEM_BASICO")
cor(data$SEM_BASICO, data$CRIMES)
var(data$SEM_BASICO, data$CRIMES, na.rm = FALSE, use)
cov(data$SEM_BASICO, data$CRIMES, use = "everything",
    method = c("pearson", "kendall", "spearman"))

print("SEM_MEDIO")
cor(data$SEM_MEDIO, data$CRIMES)
var(data$SEM_MEDIO, data$CRIMES, na.rm = FALSE, use)
cov(data$SEM_MEDIO, data$CRIMES, use = "everything",
    method = c("pearson", "kendall", "spearman"))

print("COM_SUPERIOR")
cor(data$COM_SUPERIOR, data$CRIMES)
var(data$COM_SUPERIOR, data$CRIMES, na.rm = FALSE, use)
cov(data$COM_SUPERIOR, data$CRIMES, use = "everything",
    method = c("pearson", "kendall", "spearman"))

print("DESEMPREGADOS")
cor(data$DESEMPREGADOS, data$CRIMES)
var(data$DESEMPREGADOS, data$CRIMES, na.rm = FALSE, use)
cov(data$DESEMPREGADOS, data$CRIMES, use = "everything",
    method = c("pearson", "kendall", "spearman"))
```
