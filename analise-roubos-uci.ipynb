```r
# Importação da biblioteca padrão ggplot2
library(ggplot2) 
```

```r
# Leitura dos dados
data <- read.csv("DADOS-ANALISE-ROUBOS-UCI.csv")

print(paste("O número de colunas é: ", ncol(data))) # deve ser 7
print(paste("O número de linhas é: ", nrow(data)))  # deve ser 2128

print(paste("O número de colunas é: ", ncol(data)))
print(paste("O número de linhas é: ", nrow(data)))

# imprimir início do data frame
head(data)
```

```r
print("Diagrama de dispersão")
print("[Populacao] X [% Roubos / 100 mil habitantes]")

ggplot(data=data, aes(x=populacao, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = 'Populacao', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()
```

```r
print("Diagrama de dispersão")
print("[Indice de Renda per capta] X [% Roubos / 100 mil habitantes]")

ggplot(data=data, aes(x=indice_renda_per_capta, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = 'Indice de Renda per capta', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()
```
