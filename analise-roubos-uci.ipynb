```r
# Importação da biblioteca padrão ggplot2
library(ggplot2) 
```

```r
# Leitura dos dados
csv_content <- read.csv("DADOS-ANALISE-ROUBOS-UCI.csv")

porcentagem_de_roubos_a_cada_100_mil = csv_content["porcentagem_de_roubos_a_cada_100_mil"]

populacao = csv_content["populacao"]
indice_renda_per_capta = csv_content["indice_renda_per_capta"]
porcentagem_nivel_basico = csv_content["porcentagem_nivel_basico"]
porcentagem_sem_nivel_medio = csv_content["porcentagem_sem_nivel_medio"]
porcentagem_com_nivel_superior = csv_content["porcentagem_com_nivel_superior"]
porcentagem_desempregados = csv_content["porcentagem_desempregados"]

print(paste("O número de colunas é: ", ncol(csv_content))) # deve ser 7
print(paste("O número de linhas é: ", nrow(csv_content)))  # deve ser 2128

print(paste("O número de colunas é: ", ncol(csv_content)))
print(paste("O número de linhas é: ", nrow(csv_content)))


data = data.frame(
    porcentagem_de_roubos_a_cada_100_mil = log(porcentagem_de_roubos_a_cada_100_mil),
    populacao = log(populacao),
    indice_renda_per_capta = log(indice_renda_per_capta),
    porcentagem_nivel_basico = porcentagem_nivel_basico,
    porcentagem_sem_nivel_medio = porcentagem_sem_nivel_medio,
    porcentagem_com_nivel_superior = porcentagem_com_nivel_superior,
    porcentagem_desempregados = porcentagem_desempregados
)

# imprimir início do data frame
head(data)
```

```r
print("Diagrama de dispersão")
print("[Populacao] X [% Roubos / 100 mil habitantes]")

g1 <- ggplot(data=data, aes(x=populacao, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = 'Populacao', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()
g1 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Indice de Renda per capta] X [% Roubos / 100 mil habitantes]")

g2 <- ggplot(data=data, aes(x=indice_renda_per_capta, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = 'Indice de Renda per capta', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()

g2 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Pessoas com nível básico incompleto] X [% Roubos / 100 mil habitantes]")

g3 <- ggplot(data=data, aes(x=porcentagem_nivel_basico, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = '% Pessoas com nível básico incompleto', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()
g3 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Pessoas com nível médio incompleto] X [% Roubos / 100 mil habitantes]")

g4 <- ggplot(data=data, aes(x=porcentagem_sem_nivel_medio, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = '% Pessoas com nível médio incompleto', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()

g4 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Pessoas com nível superior completo] X [% Roubos / 100 mil habitantes]")

g5 <- ggplot(data=data, aes(x=porcentagem_com_nivel_superior, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = '% Pessoas com nível superior completo', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()

g5 + geom_smooth(method = 'lm', se = F)
```

```r
print("Diagrama de dispersão")
print("[Desempregados] X [% Roubos / 100 mil habitantes]")

g6 <- ggplot(data=data, aes(x=porcentagem_desempregados, y=porcentagem_de_roubos_a_cada_100_mil))+
  geom_point() +
  labs(x = '% Desempregados', y = '% Roubos / 100 mil habitantes') +
  theme_minimal()

g6 + geom_smooth(method = 'lm', se = F)
```
