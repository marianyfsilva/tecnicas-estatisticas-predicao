```r
# Importação da biblioteca padrão ggplot2
library(ggplot2) 
# Importação da biblioteca padrão moments
library(moments)
```

```r
# Leitura dos dados
csv_content <- read.csv("DADOS-CRIMES-UCI.csv")

print(paste("O número de colunas é: ", ncol(csv_content))) # deve ser 7
print(paste("O número de linhas é: ", nrow(csv_content)))  # deve ser 2128

print(paste("O número de colunas é: ", ncol(csv_content)))
print(paste("O número de linhas é: ", nrow(csv_content)))

data = data.frame(
    # Variável resposta
    CRIMES = log(csv_content["CRIMES"]),
    # Variaveis explicativas
    POPULACAO = log(csv_content["POPULACAO"]),
    RENDA = log(csv_content["RENDA"]),
    SEM_BASICO = csv_content["SEM_BASICO"],
    SEM_MEDIO = csv_content["SEM_MEDIO"],
    COM_SUPERIOR = csv_content["COM_SUPERIOR"],
    DESEMPREGADOS = csv_content["DESEMPREGADOS"]
)

# imprimir início do data frame
head(data)
```

```r
print("VE: CRIMES")

print("Média")
mean(data$CRIMES)

print("Mediana")
median(data$CRIMES)

print("Desvio Padrão")
sd(data$CRIMES)

print("Coeficiente de Assimetria")
skewness(data$CRIMES)

print("Coeficiente de Curtose")
kurtosis(data$CRIMES)
```

```r
print("Diagrama de dispersão")
print("[Populacao] X [Total de crimes violentos e não violentos]")

g1 <- ggplot(data=data, aes(x=POPULACAO, y=CRIMES))+
  geom_point() +
  labs(x = 'Populacao', y = 'Total de crimes violentos e não violentos') +
  theme_minimal()
g1 + geom_smooth(method = 'lm', se = F)
```

```r
# Coeficiente de correlação
 
print("POPULACAO")
cor(data$POPULACAO, data$CRIMES)

print("RENDA")
cor(data$RENDA, data$CRIMES)
 
print("SEM_BASICO")
cor(data$SEM_BASICO, data$CRIMES)
 
print("SEM_MEDIO")
cor(data$SEM_MEDIO, data$CRIMES)
 
print("COM_SUPERIOR")
cor(data$COM_SUPERIOR, data$CRIMES)
 
print("DESEMPREGADOS")
cor(data$DESEMPREGADOS, data$CRIMES)

# Reta de mínimos quadrados

# Análise de Resíduos
```

```r
# Analise dos coeficientes calculados

# Ajuste do modelo de regressão linear múltipla
## Y variavel resposta ~ X1 + X2 + ... Xn todas as variaveis explicativas
modelo <- lm(CRIMES ~ POPULACAO + RENDA + SEM_BASICO + SEM_MEDIO + COM_SUPERIOR + DESEMPREGADOS, # após o '~' adicionamos todas as variáveis explicativas separadas por '+'
             data=data) # nome do objeto onde encontram-se as variáveis

summary(modelo) # imprimir coeficientes do modelo
```

```r
# ANÁLISE DE RESÍDUOS

ggplot(data = modelo) + 
  geom_point(aes(x=.fitted, 
                 y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'Valores preditos', 
       y = 'Resíduos padronizados') + 
  theme_minimal()

ggplot(data = modelo, 
       aes(sample = .stdresid)) + 
  stat_qq() + 
  stat_qq_line() + 
  labs(x = 'Valores esperados pela normal', 
       y = 'Resíduos padronizados') +
  theme_minimal()

ggplot(data = modelo) + 
  geom_histogram(aes(x = .stdresid),
                 bins = 5,
                 fill = 'lightgrey',
                 colour = 'black') + 
  labs(x = 'Resíduos padronizados', 
       y = 'Frequência') +
  theme_minimal()

# TESTE DE SHAPIRO-WILK
shapiro.test(rstandard(modelo))
```

```r

print(">>")
# Variável explicativa 1 (POPULACAO) versus Resíduos padronizados
ggplot(data = modelo) + 
  geom_point(aes(x=POPULACAO, y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'POPULACAO', y = 'Resíduos padronizados') + 
  theme_minimal()

print(">>")
# Variável explicativa 2 (RENDA) versus Resíduos padronizados
ggplot(data = modelo) + 
  geom_point(aes(x=RENDA, y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'RENDA', y = 'Resíduos padronizados') + 
  theme_minimal()

print(">>")
# Variável explicativa 3 (SEM_BASICO) versus Resíduos padronizados
ggplot(data = modelo) + 
  geom_point(aes(x=RENDA, y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'SEM_BASICO', y = 'Resíduos padronizados') + 
  theme_minimal()

print(">>")
# Variável explicativa 4 (SEM_MEDIO) versus Resíduos padronizados
ggplot(data = modelo) + 
  geom_point(aes(x=RENDA, y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'SEM_MEDIO', y = 'Resíduos padronizados') + 
  theme_minimal()

print(">>")
# Variável explicativa 5 (COM_SUPERIOR) versus Resíduos padronizados
ggplot(data = modelo) + 
  geom_point(aes(x=RENDA, y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'COM_SUPERIOR', y = 'Resíduos padronizados') + 
  theme_minimal()

print(">>")

# Variável explicativa 6 (DESEMPREGADOS) versus Resíduos padronizados
ggplot(data = modelo) + 
  geom_point(aes(x=RENDA, y=.stdresid)) +
  geom_hline(yintercept = 0) +
  labs(x = 'DESEMPREGADOS', y = 'Resíduos padronizados') + 
  theme_minimal()
```
